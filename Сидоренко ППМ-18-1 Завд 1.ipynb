__Сумма первых 1000 натуральных чисел.__
1. Встроенная функция sum() от генератора первых 1000 натуральных чисел.

```python
sum((i for i in range(1, 1001)))
```

2. Цикл.

```python
sum = 0
for i in range(1, 1001): sum += i
sum
```

3. Функция reduce() из модуля functools и генератор списков.

```python
from functools import reduce
reduce(lambda x, y: x + y, [i for i in range(1, 1001)])
```
