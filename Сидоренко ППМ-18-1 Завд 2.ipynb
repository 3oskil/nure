__2) Знак числа__

```python
x = int(input())
if x > 0:
    print('+1')
elif x < 0:
    print('-1')
else:
    print(0)
```

__3) Шахматная доска__

```python
x1, y1, x2, y2 = int(input()), int(input()), int(input()), int(input())
print('YES' if (x1 + y1 + x2 + y2) % 2 == 0 else 'NO')
```

__4) Високосный год__

```python
year = int(input())
print('YES' if year % 4 == 0 and year % 100 != 0 or year % 400 == 0 else 'NO')
```

__5) Минимум из трех чисел__

```python
a, b, c = int(input()), int(input()), int(input())
if a > b and a > c:
    print(a)
elif b > a and b > c:
    print(b)
else:
    print(c)
```

__6) Сколько совпадает чисел__

```python
a, b, c = int(input()), int(input()), int(input())
if a == b and b == c:
    print(3)
elif a != b and b != c and a != c:
    print(0)
else:
    print(2)
```

__7) Ход ладьи__

__8) Ход короля__

```python
x1, y1, x2, y2 = int(input()), int(input()), int(input()), int(input())
print('YES' if abs(x1 - x2) + abs(y1 - y2) <= 2 else 'NO')
```

__9) Ход слона__

```python
x1, y1, x2, y2 = int(input()), int(input()), int(input()), int(input())
print('YES' if abs(x1 - x2) == abs(y1 - y2) else 'NO')
```

__10) Ход ферзя__

```python
x1, y1, x2, y2 = int(input()), int(input()), int(input()), int(input())
print('YES' if abs(x1 - x2) == abs(y1 - y2) or x1 == x2 or y1 == y2 else 'NO')
```

__11) Ход коня__

```python
x1, y1, x2, y2 = int(input()), int(input()), int(input()), int(input())
print('YES' if abs(x1 - x2) == 1 and abs(y1 - y2) == 2 or abs(x1 - x2) == 2 and abs(y1 - y2) == 1 else 'NO')
```

__12) Шоколадка__

```python
n, m, k = int(input()), int(input()), int(input())
print('YES' if k < n * m and ((k % n == 0) or (k % m == 0)) else 'NO')
```

__13) Яша плавает в бассейне__

```python
n, m, x, y = int(input()), int(input()), int(input()), int(input())
if n > m:
    n, m = m, n
if x >= n / 2:
    x = n - x
if y >= m / 2:
    y = m - y
print(x if x < y else y)
```
