# Цикл for

1) Сумма факториалов

1. С рекурсивной функцией факториала

```python
n = int(input())


def factorial(n):
    return 1 if n == 0 else factorial(n - 1) * n


result = 0
for i in range(1, n + 1):
    result += factorial(i)
print(result)
```

2. С одним циклом

```python
partial_factorial = 1
partial_sum = 0
for i in range(1, int(input()) + 1):
    partial_factorial *= i
    partial_sum += partial_factorial
print(partial_sum)
```

2) Количество нулей

```python
zeros = 0
for i in range(int(input())):
    if int(input()) == 0:
        zeros += 1
print(zeros)
```

3) Лесенка

```python
for i in range(1, int(input()) + 1):
    for j in range(1, i + 1):
        print(j, sep='', end='')
    print()
```

4) Потерянная карточка

```python
n = int(input())
sum = 0
for i in range(1, n + 1):
    sum += i
for i in range(n - 1):
    sum -= int(input())
print(sum)
```

# Цикл while

1) Числа Фибоначчи

```python
n = int(input())
if n == 0:
    print(0)
else:
    a, b = 0, 1
    for i in range(2, n + 1):
        a, b = b, a + b
    print(b)
```

2) Номер числа Фибоначчи

```python
a = int(input())
if a == 0:
    print(0)
else:
    fib_prev, fib_next = 0, 1
    n = 1
    while fib_next <= a:
        if fib_next == a:
            print(n)
            break
        fib_prev, fib_next = fib_next, fib_prev + fib_next
        n += 1
    else:
        print(-1)
```

3) Максимальное число идущих подряд равных элементов

```python
prev = -1
curr_rep_len = 0
max_rep_len = 0
element = int(input())
while element != 0:
    if prev == element:
        curr_rep_len += 1
    else:
        prev = element
        max_rep_len = max(max_rep_len, curr_rep_len)
        curr_rep_len = 1
    element = int(input())
max_rep_len = max(max_rep_len, curr_rep_len)
print(max_rep_len)
```

4) Стандартное отклонение

```python
from math import sqrt
 
partial_sum = 0
partial_sum_squares = 0
x_i = int(input())
n = 0
while x_i != 0:
    n += 1
    partial_sum += x_i
    partial_sum_squares += x_i ** 2
    x_i = int(input())
print(sqrt((partial_sum_squares - partial_sum ** 2 / n) / (n - 1)))
```

# Списки

1) Количество совпадающих пар

```python
a = [int(s) for s in input().split()]
counter = 0
for i in range(len(a)):
    for j in range(i + 1, len(a)):
        if a[i] == a[j]:
            counter += 1
print(counter)
```

2) Уникальные элементы

```python
a = [int(s) for s in input().split()]
for i in range(len(a)):
    for j in range(len(a)):
        if i != j and a[i] == a[j]:
            break
    else:
        print(a[i], end=' ')
```

3) Кегельбан

```python
n, k = [int(s) for s in input().split()]
bahn = ['I'] * n
for i in range(k):
    left, right = [int(s) for s in input().split()]
    for j in range(left - 1, right):
        bahn[j] = '.'
print(''.join(bahn))
```

4) Ферзи

```python
n = 8
x = []
y = []
for i in range(n):
    new_x, new_y = [int(s) for s in input().split()]
    x.append(new_x)
    y.append(new_y)
 
correct = True
for i in range(n):
    for j in range(i + 1, n):
        if x[i] == x[j] or y[i] == y[j] or abs(x[i] - x[j]) == abs(y[i] - y[j]):
            correct = False
 
if correct:
    print('NO')
else:
    print('YES')
```
